```python
# reading the data from the file
with open('filtered_signal_dict.txt', 'rb') as handle:
    data = handle.read()
  
print("Data type before reconstruction : ", type(data))
  
# reconstructing the data as dictionary
filteredSignalDict = pickle.loads(data)
  
print("Data type after reconstruction : ", type(filteredSignalDict))
```

```python
win_len = 635
overlap=0.5
N_samp=2100
overlap = np.round(overlap*win_len)
N_samp = N_samp - win_len + 1

idx_start = np.round(np.arange(0,N_samp, win_len-overlap)).astype(int)
idx_stop = np.round(idx_start + win_len - 1)

```

```python
trainFileList=fileList[:500]
validFileList=fileList[500:600]
testFileList=fileList[600:]


len(trainFileList),len(testFileList)
```

```python
train_filtered_windowed_signal=[]
train_label=[]
test_filtered_windowed_signal=[]
test_label=[]
valid_filtered_windowed_signal=[]
valid_label=[]

N_win = len(idx_start)

for i in trainFileList:
    # print(i)
    id=obj.getIdFromFilename(i)
    SBP=SBPDict[int(id)]
    DBP=DBPDict[int(id)]
    PPG=np.squeeze(filteredSignalDict[i])
    # print(PPG.shape)
   

    for j in range(0, N_win):
        start_id = idx_start[j]
        stop_id = idx_stop[j]
        ppg_win = PPG[start_id:stop_id+1]
        train_filtered_windowed_signal.append(ppg_win)
        train_label.append(np.array([SBP,DBP]))


for i in testFileList:
    # print(i)
    id=obj.getIdFromFilename(i)
    SBP=SBPDict[int(id)]
    DBP=DBPDict[int(id)]
    PPG=np.squeeze(filteredSignalDict[i])
    # print(PPG.shape)
   

    for j in range(0, N_win):
        start_id = idx_start[j]
        stop_id = idx_stop[j]
        ppg_win = PPG[start_id:stop_id+1]
        test_filtered_windowed_signal.append(ppg_win)
        test_label.append(np.array([SBP,DBP]))


for i in validFileList:
    # print(i)
    id=obj.getIdFromFilename(i)
    SBP=SBPDict[int(id)]
    DBP=DBPDict[int(id)]
    PPG=np.squeeze(filteredSignalDict[i])
    # print(PPG.shape)
   

    for j in range(0, N_win):
        start_id = idx_start[j]
        stop_id = idx_stop[j]
        ppg_win = PPG[start_id:stop_id+1]
        valid_filtered_windowed_signal.append(ppg_win)
        valid_label.append(np.array([SBP,DBP]))
    
    # print(id)
    # print(type(id))
    # break
```
